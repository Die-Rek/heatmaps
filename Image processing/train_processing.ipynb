```python
import os
import moviepy.editor as mp
```

```python

from_directory = 'C:/Users/senne.dierick/Documents/Bachelorproef/VSCodeFolder/unprocessed_videos/'
to_directory = 'C:/Users/senne.dierick/Documents/Bachelorproef/VSCodeFolder/processed_videos/'

def new_name(filename: str):
    filename = filename.rstrip(".mp4")
    filename += "_resized.mp4"
    return filename

for filename in os.listdir(from_directory):
    if not new_name(filename) in os.listdir(from_directory):    
        clip = mp.VideoFileClip(from_directory + filename)
        clip = clip.set_fps(20)
        clip_resized = clip.resize(height=720) # make the height 720px ( According to moviePy documenation The width is then computed so that the width/height ratio is conserved.)
        file = new_name(filename)
        clip_resized.write_videofile(to_directory + file)
```
